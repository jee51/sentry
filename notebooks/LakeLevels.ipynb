# Suivi du niveau d'un lac.

```python
import json
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (15,8)
```

```python
with open("../lakes/Leman_202211191010_ALTI_V2.1.0.json") as jdf:
    jd = json.load(jdf)
```

```python
df = pd.DataFrame(jd['data']).set_index('datetime')
df
```

```python
fig,ax = plt.subplots()
x = df.index
y = df['water_surface_height_above_reference_datum']
e = df['water_surface_height_uncertainty']
y.plot(style='*-.')
ax.fill_between(x,y-e,y+e,color='b',alpha=0.2)
plt.title(jd['properties']['long_name']+' - ('+jd['properties']['lake']+')',fontsize=15)
plt.ylabel('m')
```
